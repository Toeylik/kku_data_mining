```
nb = pd.read_csv('/content/drive/My Drive/DWDM_material/newbooks.csv')
nb
```

```
nb.isnull().any()
```

```
ratings = pd.read_csv('/content/drive/My Drive/DWDM_material/ratings.csv')
ratings
```

```
ratings.isnull().any()
```

```
bookmerge = nb.merge(ratings,how='left', left_on='book_id',right_on='book_id')
bookmerge
```

```
bookmerge = bookmerge.dropna()
bookmerge
```

```
bookmerge.drop(columns=['language_code','original_publication_year','medium_image_url','small_image_url','desc','genre','image_url'])
```

```
table = bookmerge.groupby('user_id')['original_title'].apply(list)
table
```
